```python
#Import Libraries.
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
```

```python
data = load_diabetes()
```

```python
#Split the data
x= data.data
y= data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=4 ,test_size=0.2)
```

```python
#To set parameters.SGD+Ridge
reg = SGDRegressor(penalty='l2', max_iter=500, learning_rate='constant', eta0=0.1, alpha=0.001)
```

```python
#Train model
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
print("R2 Score is :",r2_score(y_test,y_pred))
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
class MYSGDRegressor:
  def __init__(self,learning_rate,alpha,epochs):
    self.learning_rate = learning_rate
    self.alpha = alpha
    self.epochs= epochs
    self.coef_ = None
    self.intercept_ = None

  def fit(self,x_train,y_train):
    self.coef_ = np.ones(x_train.shape[1])
    self.intercept_ = 0

    thetha = np.insert(self.coef_,0,self.intercept_) #Make a matrix and add 0 at position 0 of matrix which is basically self.coef_ matrix.

    x_train = np.insert(x_train,0,1,axis=1)

    for i in range(self.epochs):
      thetha_der = np.dot(x_train.T,x_train).dot(thetha) - np.dot(x_train.T,y_train) + self.alpha * thetha #Step 2
      thetha = thetha - self.learning_rate * thetha_der #step1 of this loop.
    self.coef_ = thetha[1:]
    self.intercept_ = thetha[0]

  def predict(self,x_test):
    return np.dot(x_test,self.coef_) + self.intercept_
```

```python
reg= MYSGDRegressor(learning_rate=0.005,alpha=0.001,epochs=500)
reg
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
print("R2 score is :",r2_score(y_test,y_pred))
```
